```python
import pandas as pd
from requests import get

def download(url, file_name = None):
    if not file_name:
        file_name = url.split('/')[-1]
	
    with open(file_name, "wb") as file:   
        response = get(url)               
        file.write(response.content)

data = pd.read_csv('blast.csv')
print(data)
data = data['EventID']

net = 'network=BK&'
sta = 'station=WENL&'
cha = 'channel=HHZ&'
url = 'http://service.ncedc.org/ncedcws/eventdata/1/query?'
url = url + net + sta + cha
catalog = 'catalog=NCSS&eventid='

i = 0
for id in data:
    i += 1
    download(url+catalog+str(id), str(id)+'.mseed')
    print(str(i) + ' / ' + str(len(data)) + '\n')
```
